## Definición del problema:
- Base de datos obtenida de: https://www.datos.gov.co/Salud-y-Protecci-n-Social/Base-de-Datos-Sisben-2017-Municipio-de-Medell-n/cnr8-armb/data 

Esta base de datos contiene la información anonimizada de la población sisbenizada en el Municipio de Medellín con corte a 15 de diciembre de 2017. 
Es el sistema de información colombiano que permite identificar a la población potencial beneficiaria de programas sociales. El Departamento Nacional de Planeación se encarga de implementar y administrar el instrumento SISBEN, y suministrar la información recolectada a los encargados de operar los programas sociales.

### Modelamiento de Datos

```python
#Dividir la base de datos en datos de entrenamiento y datos de prueba para aplicar los metodos supervisados.
cols = datos.columns.values
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(datos[cols[:-1]],datos['PUNTAJE'],train_size=0.5,random_state=42)
```

```python
#Importar librerias para aplicar metodos de evaluación de los algoritmos que se implementarán.
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
```

```python
#importar librerias de sklearn para metodos de predicción. 
from sklearn import linear_model
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
```

```python
b = MLPRegressor()
b.fit(Xtrain,ytrain) # Se entrena el método implementado
y_pred=b.predict(Xtest) #Se realiza una predicción
scorer = r2_score(ytest, y_pred) 
print ('r2_score: ',scorer)
#scores = cross_val_score(b, Xtrain, ytrain, cv=10)
#s = np.mean(np.sqrt(abs(scores)))
#print ('Cross-Validation: ',s)

plt.figure(figsize=(15,10))
plt.hist(y_pred,alpha=0.4,color='green')
plt.hist(ytest,alpha=0.4)
plt.title("Redes Neuronales")
```

```python
met = [linear_model.LinearRegression(),DecisionTreeRegressor(max_depth=19,random_state=1),
       RandomForestRegressor(max_depth=15,random_state=1)]
m = ['Regresion Lineal','Decision Tree','Random Forest']
for i in range (0,len(met)):
    met[i].fit(Xtrain,ytrain) # Se entrena el modelo 
    y_pred=met[i].predict(Xtest) #Se evalua el modelo
    print ('Método: ',i+1,met[i])
    scorer = r2_score(ytest, y_pred) 
    print ('r2_score: ',scorer)
    scores = cross_val_score(met[i], Xtrain, ytrain, cv=10)
    s = np.mean(np.sqrt(abs(scores)))
    print ('Cross-Validation: ',s)
    print ('\n')
    
    plt.figure(figsize=(15,10))
    plt.subplot(2,2,i+1)
    plt.hist(y_pred,alpha=0.4,color='green')
    plt.hist(ytest,alpha=0.4)
    a = m[i]
    plt.title(a)
```

### Hyperparameter tuning para Decision Tree Regressor

```python
#Importar libreria para aplicar hiperparametrización.
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'max_depth': [3, 19, 25],
    'n_estimators': [5, 10, 15]
}
model = RandomForestRegressor()
grid_search = GridSearchCV(estimator = model, param_grid = param_grid,cv = 3)
grid_search.fit(Xtrain,ytrain)
print(grid_search.best_params_)
```

```python
#Predecir con el modelo luego de hiperparametrizar.
model = grid_search.best_estimator_
yfit = model.predict(Xtest)
```

```python
scorer = r2_score(ytest, yfit) 
print ('r2_score: ',scorer)
```
